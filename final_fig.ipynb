```python
import os
import time
import toml
import jax
from jax import block_until_ready
import jax.numpy as jnp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
from jaxincell import simulation

# Force JAX to use only GPU
jax.config.update("jax_platform_name", "cpu")
print("Running on device:", jax.devices())

# Load parameters
toml_dict = toml.load('example_input_copy.toml')
base_input = toml_dict["input_parameters"]
solver_parameters = toml_dict["solver_parameters"]

def fit_growth(time, energy, t0, t1):
    mask = (time >= t0) & (time <= t1)
    t_fit = time[mask]
    y_fit = np.log(energy[mask])
    slope, intercept, _, _, _ = linregress(t_fit, y_fit)
    fit_curve = np.exp(intercept + slope * time)
    return slope, intercept, fit_curve

# Fit window
t0, t1 = 250, 450
drift_speeds = np.linspace(0.125, 0.3, 10)

# Number of pseudoelectrons to test
electron_counts = [ 500, 1000, 2000, 4000, 8000, 16000]
runtime_csv = "runtime_scaling_results_CPU.csv"

if os.path.exists(runtime_csv):
    df_runtime = pd.read_csv(runtime_csv)
    completed_runs = set(df_runtime["number_pseudoelectrons"])
else:
    df_runtime = pd.DataFrame(columns=["number_pseudoelectrons", "total_runtime_sec"])
    completed_runs = set()
    
for N in electron_counts:
    if N in completed_runs:
        print(f"Skipping N={N} (already done)")
        continue
    params = dict(base_input)
    solver_params = dict(solver_parameters)
    solver_params["number_pseudoelectrons"] = int(N)

    csv_file = f"drift_scan_results_{N}_CPU.csv"
    df_scan = pd.DataFrame(columns=["drift_speed", "growth_rate", "runtime_sec"])

    print(f"\n=== Starting run for N = {N} pseudoelectrons ===")
    total_runtime = 0.0

    for v_drift in drift_speeds:
        v_val = float(v_drift)
        params["electron_drift_speed_x"] = v_val

        start_step = time.time()
        out = block_until_ready(simulation(params, **solver_params))
        step_time = time.time() - start_step
        total_runtime += step_time

        E_field = np.array(out["electric_field_energy"])
        time_arr = np.arange(len(E_field))

        slope, intercept, fit_curve = fit_growth(time_arr, E_field, t0, t1)
        gamma = slope * 28.134884880153997 / 2

        new_row = pd.DataFrame([[v_val, gamma, step_time]], columns=["drift_speed", "growth_rate", "runtime_sec"])
        df_scan = pd.concat([df_scan, new_row], ignore_index=True)
        df_scan.to_csv(csv_file, index=False)

        print(f"  Drift={v_val:.3f} → γ={gamma:.5f}, step runtime={step_time:.2f}s")

    print(f"Completed N={N} scan in {total_runtime/60:.2f} minutes (sum of per-drift times)")

    df_runtime = pd.concat(
        [df_runtime, pd.DataFrame([[N, total_runtime]], columns=["number_pseudoelectrons", "total_runtime_sec"])],
        ignore_index=True,
    )
    df_runtime.to_csv(runtime_csv, index=False)

# Plot runtime scaling
plt.figure(figsize=(7,5))
plt.plot(df_runtime["number_pseudoelectrons"], df_runtime["total_runtime_sec"], "o-", lw=2)
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Number of pseudoelectrons (log scale)")
plt.ylabel("Total runtime (s, log scale)")
plt.title("Two-stream dispersion runtime scaling on GPU")
plt.grid(True, which="both", alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import least_squares
from scipy.special import erfi
import pandas as pd

# Plasma dispersion function
def Z(xi):
    return 1j * np.sqrt(np.pi) * np.exp(-xi**2) * (1 + 1j * erfi(xi))

# Symmetric two-stream dispersion relation
def epsilon(omega, k, v_b, v_th=1.0):
    xi1 = (1j*omega*v_th - k*v_b) / (np.sqrt(2)*k*v_th)
    xi2 = (1j*omega*v_th + k*v_b) / (np.sqrt(2)*k*v_th)
    return 1 + 0.5*(1/(k**2)) * (2 + xi1*Z(xi1) + xi2*Z(xi2))

# Wrapper for least_squares solver
def func_for_solver(omega_parts, k, v_b, v_th=1.0):
    omega = omega_parts[0] + 1j*omega_parts[1]
    val = epsilon(omega, k, v_b, v_th)
    return [val.real, val.imag]

# Analytic solution
k = 0.125
v_th = 0.05 / np.sqrt(2)
vb_values = np.linspace(0.125, 0.3, 50)

gamma_values_analytic = []
guess = [0.01, 0.01]

for vb in vb_values:
    res = least_squares(func_for_solver, guess, args=(k, vb, v_th), gtol=2.2e-25, max_nfev=10000)
    omega_root = res.x[0] + 1j*res.x[1]
    gamma_values_analytic.append(omega_root.real)
    guess = res.x  # use previous solution as next guess

# List of specific simulation files to plot
sim_files = [
    "drift_scan_results_500_GPU.csv",
    "drift_scan_results_1000_GPU.csv",
    "drift_scan_results_2000_GPU.csv",
    "drift_scan_results_4000_GPU.csv",
    "drift_scan_results_8000_GPU.csv",
    "drift_scan_results_16000_GPU.csv"
]

plt.figure(figsize=(8,5))
colors = ["C1", "C2", "C3", "C4", "C5", "C6"]  # Distinct colors for each simulation

for csv_file, color in zip(sim_files, colors):
    df = pd.read_csv(csv_file)
    N_label = csv_file.replace("drift_scan_results_", "").replace(".csv", "")
    plt.plot(df["drift_speed"], df["growth_rate"], "o--", color=color, label=f"Sim N={N_label}")

# Plot analytic curve
plt.plot(vb_values, gamma_values_analytic, "s-", color="C0", linewidth=2, label="Analytic (Plasma Dispersion)")

plt.xlabel(r"Beam velocity $v_b$")
plt.ylabel(r"Growth rate $\gamma$")
plt.title("Two-stream instability: Growth rate vs Beam Velocity")
plt.grid(True, which="both", alpha=0.4)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from jaxincell import plot
from jaxincell import simulation
from jax import block_until_ready

input_parameters = {
"length"                       : 3e-1,    # dimensions of the simulation box in (x, y, z)
"amplitude_perturbation_x"     : 5e-9,       # amplitude of sinusoidal perturbation in x
"wavenumber_electrons_x"       : 1,       # wavenumber of perturbation in z
"grid_points_per_Debye_length" : 0.7,     # dx over Debye length
"velocity_plus_minus_electrons_z": False, # create two groups of electrons moving in opposite directions
"velocity_plus_minus_electrons_x": False, # create two groups of electrons moving in opposite directions
"random_positions_x": True,  # Use random positions in x for particles
"random_positions_y": True,  # Use random positions in y for particles
"random_positions_z": True,  # Use random positions in z for particles
"electron_drift_speed_x": 0, # Drift speed of electrons in x direction
"electron_drift_speed_z": 0, # Drift speed of electrons in z direction
"ion_temperature_over_electron_temperature_x": 1, #not 10!, # Temperature of ions over temperature of electrons
"print_info"                   : True,    # print information about the simulation
"vth_electrons_over_c_x": 0.01,           # Thermal velocity of electrons over speed of light
"vth_electrons_over_c_z": 0.10,           # Thermal velocity of electrons over speed of light
"timestep_over_spatialstep_times_c": 0.5, # dt * speed_of_light / dx
"relativistic": False,  # Use relativistic equations of motion
}

solver_parameters = {
    "field_solver"           : 0,    # Algorithm to solve E and B fields - 0: Curl_EB, 1: Gauss_1D_FFT, 2: Gauss_1D_Cartesian, 3: Poisson_1D_FFT, 
    "number_grid_points"     : 201,  # Number of grid points
    "number_pseudoelectrons" : 3000, # Number of pseudoelectrons
    "total_steps"            : 6000, # Total number of time steps
    "time_evolution_algorithm": 0,  # Algorithm to evolve particles in time - 0: Boris, 1: Implicit_Crank Nicholson (electrostatic)
}

output = block_until_ready(simulation(input_parameters, **solver_parameters))

plot(output, direction="xz")  # Plot the results in x and z direction
```

```python
def setup(ax):
    ax.tick_params(direction='in', length=10, width=4, colors='black')
    ax.spines['left'].set_linewidth(3)
    ax.spines['right'].set_linewidth(3)
    ax.spines['top'].set_linewidth(3)
    ax.spines['bottom'].set_linewidth(3)
    #ax.spines['right'].set_color('none')
    #ax.spines['left'].set_color('none')
    #ax.yaxis.set_major_locator(ticker.NullLocator())
    #ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('both')
    ax.yaxis.set_ticks_position('both')
    ax.tick_params(which='major', direction='in',width=3.00,length=7,labelsize=24)
    ax.tick_params(which='major', direction='in',width=3.00,length=7,labelsize=24)
    ax.tick_params(which='minor', direction='in', width=2.00,length=5)
    ax.tick_params(which='minor', direction='in',width=2.00,length=5)
    ax.patch.set_alpha(0.0)
```

```python
import jax.numpy as jnp
import matplotlib.pyplot as plt

# --- Extract data ---
E = output["electric_field_energy"]
B = output["magnetic_field_energy"]
Energy = output["total_energy"]
time = jnp.arange(len(E))

# --- Compute relative energy conservation ---
E0 = Energy[0]
rel_energy = (Energy - E0) / E0

# --- Plot section ---
fig, ax1 = plt.subplots(figsize=(10, 8))
time_o = time / 40.40610178208843

# Plot field energies
ax1.semilogy(time_o, E, label="Electric field energy", lw=3, color="C0")
ax1.semilogy(time_o, B, label="Magnetic field energy", lw=3, color="C1")
# ax1.semilogy(time_o, Energy, label="Total energy", lw=3, color="C2")
ax1.plot(time_o, rel_energy, "C3", lw=3, label="Relative energy drift")
ax1.set_xlabel("Time ($\\omega_{pe}^{-1}$)", fontsize=24)
ax1.set_ylabel("Energy (a.u.)", fontsize=24)
ax1.legend(fontsize=20)


# --- Apply your style and layout ---
setup(ax1)
fig.tight_layout()
plt.show()
```
